```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps

%matplotlib inline
```

```python
n = 10000

def UniformEstimates(theta, estimates):    
    #Генерируем выборку
    sample = sps.uniform.rvs(0, theta, n)
    
    #Заполняем массивы соответсвующими оценками параметра theta
    estimates[0] = np.array([2 * sample[:i].mean() 
                            for i in range(1, n + 1)])
    estimates[1] = np.array([sample[:i].mean() + sample[:i].max() / 2 
                            for i in range(1, n + 1)])
    estimates[2] = np.array([(i + 1) * sample[:i].min() 
                            for i in range(1, n + 1)])
    estimates[3] = np.array([sample[:i].min() + sample[:i].max() 
                            for i in range(1, n + 1)])
    estimates[4] = np.array([sample[:i].max() * (i + 1) / i 
                            for i in range(1, n + 1)])
```

```python
estimates = np.zeros((5, n))
theta = 1

UniformEstimates(theta, estimates)

#Строим график
plt.figure(figsize=(15, 8))
plt.plot(abs(estimates[0] - theta), color='red', 
             label=r'$2 \cdot \overline{X}$')
plt.plot(abs(estimates[1] - theta), color='blue', 
             label=r'$\overline{X} + \frac{X_{(n)}}{2}$')
plt.plot(abs(estimates[2] - theta), color='green', 
             label=r'$(n+1) \cdot X_{(1)}$')
plt.plot(abs(estimates[3] - theta), color='black', 
             label=r'$X_{(1)} + X_{(n)}$')
plt.plot(abs(estimates[4] - theta), color='yellow', 
             label=r'$\frac{n + 1}{n} \cdot X_{(n)}$')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), 
           ncol=3, shadow=True)
plt.show()
    
#Строим график в масштабе
plt.figure(figsize=(15, 8))
plt.ylim(0, 0.05)
plt.plot(abs(estimates[0] - theta), color='red', 
             label=r'$2 \cdot \overline{X}$')
plt.plot(abs(estimates[1] - theta), color='blue', 
             label=r'$\overline{X} + \frac{X_{(n)}}{2}$')
plt.plot(abs(estimates[3] - theta), color='black', 
             label=r'$X_{(1)} + X_{(n)}$')
plt.plot(abs(estimates[4] - theta), color='yellow', 
             label=r'$\frac{n + 1}{n} \cdot X_{(n)}$')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), 
           ncol=3, shadow=True)
plt.show()
    
print(abs(estimates[0][n - 1] - theta), abs(estimates[1][n - 1] - theta), 
      abs(estimates[2][n - 1] - theta), abs(estimates[3][n - 1] - theta), 
      abs(estimates[4][n - 1] - theta))
```
