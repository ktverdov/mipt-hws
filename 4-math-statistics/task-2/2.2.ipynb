```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import scipy.stats as sps

%matplotlib inline
```

```python
def ExponentialEstimate(k, h):
    n = 10000
    lambd = 1
    
    #Генерируем выборку
    sample = sps.expon.rvs(size=n)
    
    #Строим по выборке оценку
    exp_est = np.zeros(n)
    for i in range(1, n + 1):
        exp_est[i - 1] = (mt.factorial(k) / (sum(sample[:i] ** k) / i)) ** 
                                    (1 / float(k))
    
    #строим график
    plt.figure()
    plt.ylim(0, h)
    plt.plot(abs(exp_est - lambd), color='red', label='k = {}'.format(k))
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), 
               ncol=1, shadow=True)
    plt.show()    

def TestingArgument():
    #пробуем разные k
    #для каждого k задаем свой масштаб графика по OY
    testData = ((1, 2, 3, 4, 5, 10, 13, 15, 20, 25, 50, 75, 100),
                (0.03, 0.03, 0.03, 0.03, 0.1, 0.5, 0.5, 0.5, 2, 2, 5, 5, 5))
    
    for i in range(0, 13):
        ExponentialEstimate(testData[0][i], testData[1][i])
```

```python
TestingArgument()
```

Опыт показал, что при $k\in[1, 3]$ оценка $\left(\frac{k!}{\overline{X^k}}\right)^\frac{1}{k}$ ведет себя лучше всего.
