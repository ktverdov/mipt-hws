```python
import numpy as np
import scipy.stats as sps
from scipy.stats import pareto
from scipy.stats import cauchy
import matplotlib.pyplot as plt

%matplotlib inline
```

### Распределение Парето

```python
N = 10000

#У распределения Парето, имеющего плотность b / x**(b+1) for x >= 1
#конечны первые 4 момента, а пятый бесконечен
b = 5
sample_pareto = pareto.rvs(b, size=N)
```

```python
plt.figure(figsize=(10, 5))

#точки выборки
plt.scatter(sample_pareto, np.zeros(N) - 0.2, alpha=0.1, label='sample')

#график плотности
plt.xlim(0.9, 5.0)
grid = np.linspace(1, 5)
plt.plot(grid, pareto.pdf(grid, b), color='red', label='density')

plt.legend()
plt.show()
```

```python
#оценка дисперсии
pareto_var_est = np.array([sample_pareto[:i].var() for i in range(1, N + 1)]) 

plt.figure(figsize=(10, 5))
plt.plot(abs(pareto_var_est - pareto.var(b)))
plt.title(r'|s^2 - Var(Pareto)|')
plt.plot()
```

### Распределение Коши

```python
sample_cauchy = cauchy.rvs(size=N)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(sample_cauchy, np.zeros(N) - 0.005, alpha=0.01,label='sample')

lim_x = 5
plt.xlim(-lim_x, lim_x)
grid = np.linspace(-lim_x, lim_x)
plt.plot(grid, cauchy.pdf(grid), color='red', label='density')

plt.legend()
plt.show()
```

```python
cauchy_var_est = np.array([sample_cauchy[:i].var() for i in range(1, N + 1)]) 

plt.figure(figsize=(10, 5))
plt.plot(cauchy_var_est)
plt.title(r's^2')
plt.plot()
```

Выборочная дисперсия для распределения Парето стремится к истинному значению дисперсии, 
оценка дисперсии для Коши не имеет предела.
